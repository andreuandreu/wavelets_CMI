```python
max_val = 1.
n_steps = 1000
freq = 0.01
```

```python

t = np.arange( 0, max_val, max_val/n_steps ) 
a = np.sin(2*np.pi*freq*t + 1.5*np.pi) +1.0
```

```python
plt.plot(t, a)
```
